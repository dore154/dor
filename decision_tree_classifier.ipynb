```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('data_dec_class.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Attended_Coaching'] = le.fit_transform(df['Attended_Coaching'])  # Yes=1, No=0
df['Previous_Result'] = le.fit_transform(df['Previous_Result'])      # Pass=1, Fail=0
df['Pass'] = le.fit_transform(df['Pass'])                            # Yes=1, No=0
```

```python
X=df[['Hours_Studied','Attended_Coaching','Previous_Result']]
y=df['Pass']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn.tree import plot_tree
```

```python
from matplotlib.pyplot import rcParams
rcParams['figure.figsize']=80,50
plot_tree(clf)
```
